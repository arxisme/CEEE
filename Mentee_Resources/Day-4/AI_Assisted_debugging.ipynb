```python
### THE BROKEN CODE (DO NOT FIX HERE - JUST RUN IT TO SEE THE FAKE ACCURACY) ###
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load data
data = load_breast_cancer()
X = data.data
y = data.target

# THE BUG IS HERE! We are scaling the ENTIRE dataset before splitting it.
# This leaks information from the test set into the training set.
scaler = StandardScaler()
X_scaled_all = scaler.fit_transform(X)

# Split the already-scaled data
X_train, X_test, y_train, y_test = train_test_split(X_scaled_all, y, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)
preds = model.predict(X_test)

print(f"Flawed Model Accuracy: {accuracy_score(y_test, preds):.4f}")
print("This accuracy is a lie! The model peaked at the test data during scaling.")
```
