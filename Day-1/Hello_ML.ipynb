# Machine Learning Environment Verification
This notebook verifies that your local environment is correctly configured for Data Science, Traditional Machine Learning, and Deep Learning tasks.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import torch
import torch.nn as nn
import torch.optim as optim
print('All libraries imported successfully!')
```

## 1. Data Manipulation and Visualization (Pandas & Matplotlib)
Load the Iris dataset and visualize the feature distributions.

```python
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
plt.figure(figsize=(8, 5))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['target'], cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Dataset: Sepal Dimensions')
plt.show()
```

## 2. Traditional Machine Learning (Scikit-Learn)
Train a Random Forest classifier.

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('target', axis=1), df['target'], test_size=0.2, random_state=42)
model = RandomForestClassifier(n_estimators=10)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(f'Random Forest Accuracy: {accuracy * 100:.2f}%')
```

## 3. Deep Learning & Hardware Acceleration (PyTorch)
Verify GPU/MPS availability and train a simple Neural Network on synthetic data.

```python
# Hardware Check
if torch.cuda.is_available():
    device = torch.device('cuda')
    print('GPU (CUDA) is available.')
elif torch.backends.mps.is_available():
    device = torch.device('mps')
    print('GPU (MPS/Apple Silicon) is available.')
else:
    device = torch.device('cpu')
    print('GPU not found. Using CPU.')

# Simple Neural Network Definition
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc = nn.Linear(10, 2)
    def forward(self, x):
        return self.fc(x)

net = SimpleNN().to(device)
criterion = nn.MSELoss()
optimizer = optim.SGD(net.parameters(), lr=0.01)

# Synthetic Data
inputs = torch.randn(5, 10).to(device)
targets = torch.randn(5, 2).to(device)

# Training Step
optimizer.zero_grad()
outputs = net(inputs)
loss = criterion(outputs, targets)
loss.backward()
optimizer.step()
print(f'\nNeural Network ran successfully. Loss: {loss.item():.4f}')
```
